```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('companies.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.dropna(subset = ['name','status'], inplace= True)
df.shape
```

```python
df
```

```python

df['short_description'] = df['short_description'].fillna("Not Provided")
df
```

```python

df['description'] = df['description'].fillna("Not Provided")
```

```python
for col in ['country_code','state_code','city','region']:
      if col in df.columns:
        df[col] = df[col].fillna("unknown")
```

```python
print(df[['description','short_description','city','region']].head())
```

```python
df = df.drop(['id','Unnamed: 0.1','entity_id','parent_id'], axis = 1 )
df.head()
```

```python
df = df.drop(['permalink','created_by','normalized_name','domain','homepage_url','logo_url','logo_width','logo_height','overview','twitter_username','tag_list','state_code','first_investment_at', 'last_investment_at','first_milestone_at','last_milestone_at','invested_companies','investment_rounds'],axis = 1)
df.head()
```

```python
df.columns
```

```python
for col in ['founded_at','last_funding_at','first_funding_at','closed_at','created_at','updated_at']:
    df[col] = pd.to_datetime(df[col], errors='coerce')

print(df['founded_at'].dtypes)
print(df['founded_at'].head())
```

```python
df.head()
```

```python
for col in ['funding_total_usd','ROI']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print(df['funding_total_usd'].dtypes)
print(df['funding_total_usd'].head())
```

```python
text_columns = ['status','category_code','city']
for col in text_columns:
    df[col] = df[col].astype(str)

print(df[text_columns].dtypes)    
```

```python
df.head()
```

```python
df['lat'] = pd.to_numeric(df['lat'], errors='coerce')
df['lng'] = pd.to_numeric(df['lng'], errors='coerce')
```

```python
df['lat_invalid'] = (df['lat'].isna()) | (df['lat'] < -90) | (df['lat'] > 90)
df['lng_invalid'] = (df['lng'].isna()) | (df['lng'] < -180) | (df['lng'] > 180)

print("Invalid latitudes:", df['lat_invalid'].sum())
print("Invalid longitudes:", df['lng_invalid'].sum())

df.head()
```

```python
# Keep only a certain number of rows (e.g., 50,000)
df = df.sample(n=50000, random_state=42)  # change n as needed
```

```python
df.to_csv('final2.csv', index=False)
```
